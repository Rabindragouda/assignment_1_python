```python
##Takes two numbers as input from the user.
num_1 = int(input("Enter the number"))
num_2= int(input("Enter the number"))
print(num_1)
print(num_2)
```

```python
## Performs the basic mathematical operations on these two numbers:
#o	Addition
#o	Subtraction
#o	Multiplication
#o	Division
sum=num_1+num_2
print("Addition of two numbers",sum)
subtract = num_1-num_2
print("subtraction of two numbers",subtract)
multyply=num_1*num_2
print("Multiplication of two numbers",multyply)
division=num_1/num_2
print("Division of two numbers",division)
```
